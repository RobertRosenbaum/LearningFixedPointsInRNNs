```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import pickle
import seaborn as sns

sns.set(context='paper',style='white',font_scale=1.2,rc={"lines.linewidth":1.5})
sns.set_palette('muted')
%config InlineBackend.figure_format = 'retina'


fname='./data/MNISTFixedPoints_dW1Ratetanh.npy'



def read_objects(filename):
    with open(filename, 'rb') as f:
        objects = pickle.load(f)
    obj1, obj2 = objects

    return obj1, obj2

model, Training_Stats = read_objects(fname)


LearningRates = Training_Stats['LearningRates']
Losses = Training_Stats['TrainingLosses']
Accuracies = Training_Stats['TrainingAccuracies']
TestLosses = Training_Stats['TestLosses']
TestAccuracies = Training_Stats['TestAccuracies']
Jacobian = Training_Stats['Jacobian']


plt.subplots(figsize=(8,5))
plt.subplot(2,2,1)
for kk,lr in enumerate(LearningRates):
    plt.plot(Losses[kk].T, label=r'$\eta = $' + str(lr))
plt.xlabel("iteration")
plt.ylabel("training loss")
plt.ylim([0, 3])
plt.title('A',loc='left')
#plt.legend()
sns.despine()


plt.subplot(2,2,2)
for kk,lr in enumerate(LearningRates):
    plt.plot(TestLosses[kk].T, label=r'$\eta = $' + str(lr))
plt.xlabel("iteration")
plt.ylabel("test loss")
plt.ylim([0, 3])
plt.title('B',loc='left')
plt.legend(loc=[1.0,0])
sns.despine()

plt.subplot(2,2,3)
for kk,lr in enumerate(LearningRates):
    plt.plot(100 * Accuracies[kk].T, label=r'$\eta = $' + str(lr))
plt.xlabel("iteration")
plt.ylabel("training accuracy (%)")
plt.ylim([0, 100])
plt.title('C',loc='left')
#plt.legend()
sns.despine()

plt.subplot(2,2,4)
for kk,lr in enumerate(LearningRates):
    plt.plot(100 * TestAccuracies[kk].T, label=r'$\eta = $' + str(lr))
plt.xlabel("iteration")
plt.ylabel("test accuracy (%)")
plt.ylim([0, 100])
plt.title('D',loc='left')
#plt.legend()
sns.despine()

plt.tight_layout()

maxrealeval=np.zeros(len(LearningRates))
for kk in range(len(LearningRates)):
    maxrealeval[kk]=np.max(np.real(Jacobian[kk]))

print('max real part of evals:',maxrealeval)

plt.savefig('./Figures/MNISTdW1Ratetanh.pdf')


# # Plot Jacobian
# plt.subplots(figsize=(8,5))
# for kk in range(len(LearningRates)):
#     if len(Jacobian[kk])>0:
#         plt.plot(np.real(Jacobian[kk]),np.imag(Jacobian[kk]),'.', label=r'$\eta = $' + str(LearningRates[kk]))
#         plt.title('Eigenvalues of W after learning')
#         plt.legend()
#         plt.show()
#     else:
#         print('No eigenvalues to plot for learning rate',LearningRates[kk])
```
